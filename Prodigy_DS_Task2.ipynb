```python
df["bedrooms"]=df["bedrooms"].fillna(df["bedrooms"].mean())
df["guestroom"]=df["guestroom"].fillna(df["guestroom"].mode()[0])
df["basement"]=df["basement"].fillna(df["basement"].mode()[0])
df["hotwaterheating"]=df["hotwaterheating"].fillna(df["hotwaterheating"].mode()[0])
df["airconditioning"]=df["airconditioning"].fillna(df["airconditioning"].mode()[0])
df["prefarea"]=df["prefarea"].fillna(df["prefarea"].mode()[0])
df["parking"]=df["parking"].fillna(df["parking"].mode()[0])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'])
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['price'],df['area'])
plt.title("House price vs area")
plt.xlabel("price")
plt.ylabel("area")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='guestroom',data=df)
plt.title('count plot for guestroom')
plt.xlabel('guestroom')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='airconditioning', y='price', data=df)
plt.title('House Prices by airconditioning')
plt.xlabel('airconditioning')
plt.ylabel('Price')
plt.xticks(rotation=25)
plt.show()
```

```python
sns.pairplot(df,hue='price',diag_kind="hist")
plt.show()
```
